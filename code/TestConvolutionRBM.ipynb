```python
import os
import theano
import theano.tensor as T
import rbm
import numpy
import time
from utils import tile_raster_images
import PIL.Image as Image
import matplotlib.pyplot as pl
from logistic_sgd import load_data
from theano.tensor.shared_randomstreams import RandomStreams
```

```python
%pylab inline
```

```python
datasets = load_data('mnist.pkl.gz')
```

```python
batch_size = 50
image_shape = (28, 28)
filter_shape = (5, 5)
input_layers = 1
hidden_layers = 100
train_set_x, train_set_y = datasets[0]
test_set_x, test_set_y = datasets[2] 
n_train_batches = train_set_x.get_value(borrow=True).shape[0]/batch_size
```

### RBM

```python
persistent_chain = theano.shared(
    numpy.zeros(
        (batch_size, 500),
        dtype=theano.config.floatX
    ),
    borrow=True
)
reload(rbm)
net = rbm.RBM(n_visible=28*28, n_hidden=500)
cost, updates = net.get_cost_updates(lr=0.1, persistent=None, k=15)
```

```python
index = T.lscalar()
train_rbm = theano.function(
    [index],
    cost,
    updates=updates,
    givens={
        net.input: train_set_x[index*batch_size:(index + 1)*batch_size]
    },
    name='train_rbm'
)
```

```python
training_epochs = 15
# go through training epochs
for epoch in xrange(training_epochs):
    # go through the training set
    mean_cost = []
    for batch_index in xrange(n_train_batches):
        mean_cost += [train_rbm(batch_index)]
        if len(mean_cost) % 100 == 0:
            print np.mean(mean_cost[-100:])

    print 'Training epoch %d, cost is ' % epoch, numpy.mean(mean_cost)
```

```python
f = theano.function([], net.sample_h_given_v(train_set_x[:batch_size]))
h_raw, h_prob, h_sample = f()
f = theano.function([], net.sample_v_given_h(h_sample))
v_raw, v_prob, v_sample = f()
```

### Convolutional RBM

```python
reload(rbm)

persistent_chain = theano.shared(
    numpy.zeros((batch_size, hidden_layers, 
                 image_shape[0] - filter_shape[0] + 1, 
                 image_shape[1] - filter_shape[1] + 1),
                 dtype=theano.config.floatX),
    borrow=True
)

net = rbm.ConvolutionalRBM(image_shape, filter_shape, input_layers, hidden_layers, batch_size)
cost, updates = net.get_cost_updates(lr=1e-4, k=10, sparsity=.01, persistent=persistent_chain)
index = T.lscalar()
train_rbm = theano.function(
    [index],
    cost,
    updates=updates,
    givens={
        net.input: T.unbroadcast(train_set_x[index*batch_size:(index + 1)*batch_size].reshape((batch_size, 1, 28, 28)), 0, 1)
    },
    name='train_rbm'
)
plotting_time = 0.
start_time = time.clock()
training_epochs = 15
```

```python
# go through training epochs
for epoch in xrange(training_epochs):
    # go through the training set
    mean_cost = []
    for batch_index in xrange(n_train_batches):
        mean_cost.append(train_rbm(batch_index))
        if len(mean_cost) % 10 == 0:
            print np.mean(mean_cost[-10:])
            print '  Mean weights: ', mean(net.W.get_value())
            print '  Hidden bias: ', mean(net.hbias.get_value())

    print 'Training epoch %d, cost is ' % epoch, numpy.mean(mean_cost)

    # Construct image from the weight matrix
    pl.imshow(
        tile_raster_images(
            X=net.W.get_value(borrow=True).reshape(input_layers*hidden_layers, 5, 5),
            img_shape=(5, 5),
            tile_shape=(1, 10),
            tile_spacing=(1, 1)
        )
    )
```
